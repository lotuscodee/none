```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')

print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
grouped = df.groupby('species')

print(grouped.mean())
```

```python
print(grouped.std())
```

```python
print(grouped.quantile(0.25))

print(grouped.quantile(0.50))

print(grouped.quantile(0.75))
```

```python
for species in df['species'].unique():
    
    print("\nSPECIES:", species)
    
    print(
        df[df['species'] == species].describe()
    )
```
