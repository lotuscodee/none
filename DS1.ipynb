```python
import pandas as pd

from sklearn import datasets
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df['species'] = iris.target

print(df.head())
```

```python
    print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df['species'] = df['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})

print(df.head())
```

```python
scaler = MinMaxScaler()

df.iloc[:,0:4] = scaler.fit_transform(
    df.iloc[:,0:4]
)

print(df.head())
```

```python
encoder = LabelEncoder()

df['species'] = encoder.fit_transform(
    df['species']
)

print(df.head())
```
