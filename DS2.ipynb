```python
import pandas as pd
import numpy as np

# Create dataset
data = {
    'Name': ['Amit', 'Riya', 'Rahul', 'Sneha', 'Karan'],
    'Marks': [85, 90, None, 40, 300],
    'StudyHours': [4, 5, 3, None, 20]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
df['Marks'] = df['Marks'].fillna(
    df['Marks'].mean()
)

df['StudyHours'] = df['StudyHours'].fillna(
    df['StudyHours'].mean()
)

print(df)
```

```python
Q1 = df[['Marks', 'StudyHours']].quantile(0.25)

Q3 = df[['Marks', 'StudyHours']].quantile(0.75)

IQR = Q3 - Q1

print("IQR:")
print(IQR)
```

```python
outliers = (
    (df[['Marks','StudyHours']] < (Q1 - 1.5 * IQR)) |
    (df[['Marks','StudyHours']] > (Q3 + 1.5 * IQR))
)

print(outliers)
```

```python
median_marks = df['Marks'].median()

df.loc[df['Marks'] > 150, 'Marks'] = median_marks

print(df)
```

```python
df['Log_Marks'] = np.log(df['Marks'])

print(df)
```
