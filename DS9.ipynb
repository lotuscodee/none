```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')

print(df.head())
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df
)
```

```python
plt.title("Age Distribution by Gender and Survival")
```

```python
plt.show()
```
